```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('./data/titanic.csv')
df
```

```python
# Eliminar datos vacios
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
# Grafico de Cajas con Pandas (Matplotlib)

df.boxplot(
    column='Age', 
    by='Sex'
)
```

```python
# Grafico de Cajas con Seaborn
sns.boxplot(
    df, 
    x='Sex', 
    y='Age'
).set(title='Numero de Pasajeros por Sexo')
```

```python
# Violin Plot con Seaborn
sns.violinplot(df, x='Sex', y='Age').set(title='Numero de Pasajeros por Sexo')
```

```python
sns.violinplot(
    df, 
    x='Sex', 
    y='Age', 
    inner='quartiles' # point, stick
).set(title='Numero de Pasajeros por Sexo')
```

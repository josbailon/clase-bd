```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('./data/titanic.csv')
df
```

```python
# Eliminar datos vacios
df = df.dropna(subset=['Age'], axis='index')
df
```

```python
# KDE con Seaborn
sns.histplot(
    data=df,
    x='Age',
    kde=True,

)
```

```python
# KDE con Seaborn con Hue
sns.histplot(
    data=df,
    x='Age',
    kde=True,
    hue='Sex',
)
```
